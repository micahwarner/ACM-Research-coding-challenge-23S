```python
#check if element already is in a list
def check(list, compare):
    for x in list:
        if x == compare:
            return False
    return True

#get average of each type of data for each color of star
def printAverages(colorArr, countArr, arr):
    datas = ['Temperature', 'Luminosity', 'Radius', 'Absolute Magnitude', 'Star Type']
    i = 0
    #get each color
    while i < len(colorArr):
        n = 0
        total = [0, 0, 0, 0, 0]
        count = countArr[i]
        currentColor = colorArr[i]
        #get each star
        for star in arr[:]:
            #check if star is the desired color
            if star[5] == currentColor:
                #if it is, add data to totals
                total[0]+=star[0]
                total[1]+=star[1]
                total[2]+=star[2]
                total[3]+=star[3]
                total[4]+=star[4]
        n = 0
        #print results
        print('\033[1m' + "Averages the Color: " + str(colorArr[i]))
        print("Amount of Color: " + str(countArr[i]))
        while n < 5:
            print('\033[0m' + "-" + str(datas[n]) + ": " + str(total[n]/count))
            n+=1
        i+=1
                
#read data set
df = pd.read_csv('../input/star-dataset/6 class csv.csv')
arr = df.to_numpy()
starsColors = []
#put star colors from csv data into 2D array
for col in arr[:, 5]:
    starsColors.append(col)
colorsNoRepeat = []
colorCount = []
starsNew = [ [0]*7 for i in range(len(arr))]
i = 0
n = 0
x = len(starsColors)
#create array with no repeated colors and array with amount of stars of each color
while i < x:
    currentColor = starsColors[i].lower()
    num = 0
    lowCol = ""
    while num < len(currentColor):
        if(ord(currentColor[num]) >= 97 and ord(currentColor[num]) <= 122):
            lowCol+=(currentColor[num])
        elif(num != len(currentColor)-1):
            lowCol+=(' ')
        num+=1
    if check(colorsNoRepeat, lowCol):
        colorsNoRepeat.append(lowCol)
        colorCount.append(1)
    else:
        colorCount[colorsNoRepeat.index(lowCol)]+=1
    i+=1

#change color of star in the array to all match
for star in arr[:]:
    currentColor = star[5].lower()
    num = 0
    lowCol = ""
    while num < len(currentColor):
        if(ord(currentColor[num]) >= 97 and ord(currentColor[num]) <= 122):
            lowCol+=(currentColor[num])
        elif(num != len(currentColor)-1):
            lowCol+=(' ')
        num+=1
    star[5] = lowCol;
printAverages(colorsNoRepeat, colorCount, arr)
```
